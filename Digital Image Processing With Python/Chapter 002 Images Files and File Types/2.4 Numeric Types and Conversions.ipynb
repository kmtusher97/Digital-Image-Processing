## Converting images in Python

```python
import skimage as sk
import skimage.io as io
```

### 2.4.1 Convert to boolean

Input image should be a grayscale image

```python
img = io.imread('_images/hiramon.png')
io.imshow(img)
```

```python
b_img = sk.util.img_as_bool(img)
io.imshow(b_img)
```

### 2.4.2 Convert to 64-bit floating point

Input image can be both garyscale and RGB

```python
f64_img = sk.util.img_as_float(img)
io.imshow(f64_img)
```

### 2.4.3 Convert to 16-bit integer

```python
img = io.imread('_images/road.jpg')
io.imshow(img)
```

```python
i16_img = sk.util.img_as_int(img)
io.imshow(i16_img)
```

### 2.4.4 Convert to 8-bit unsigned integer

```python
i8ub_img = sk.util.img_as_ubyte(img)
io.imshow(i8ub_img)
```

### 2.4.5 Convert to 16-bit unsigned integer

```python
i16ub_img = sk.util.img_as_uint(img)
io.imshow(i16ub_img)
```
