## Grayscale Images


```python
import skimage.io as io

img = io.imread('_images/caribou.jpg')
```

### Convert into ``double``

```python
img_d1 = img.astype(float)
```

```python
import skimage as sk
```

```python
img_d2 = sk.util.img_as_float(img)
```

### Check the data type differences of the images

```python
img.dtype, img_d1.dtype, img_d2.dtype
```

### Display the images

```python
io.imshow(img)
```

```python
io.imshow(img_d1)
```

```python
io.imshow(img_d2)
```

```python
io.imshow(img/512.0)
```

```python
io.imshow(img/128.0)
```

```python
io.imshow(img/2.0, vmin=0, vmax=255)
```

```python
io.imshow(img/6.0, vmin=0, vmax=255)
```

```python
io.imshow(img/0.2, vmin=0, vmax=255)
```

## Binary images

### Converting into logical/binary image

```python
img_b = img > 150
img_b.dtype
```

```python
io.imshow(img_b)
```

```python
img_b = sk.util.img_as_ubyte(img_b)
io.imshow(img_b)
```

```python
import numpy as np
```

```python
img_b = np.uint8(img_b * 255)
io.imshow(img_b)
```

```python
img_b = sk.util.img_as_bool(img)
img_b = np.float16(img_b * 1)
io.imshow(img_b)
```

```python
img_b = sk.util.img_as_bool(img)
io.imshow(img_b)
```
