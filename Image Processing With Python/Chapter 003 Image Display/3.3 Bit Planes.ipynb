### 3.3.1 The bit planes of an 8bit grayscale image

get the the images with bits(0-7)

```python
bit_img = [(img >> i) % 2 for i in range(8)]
io.imshow(bit_img[3])
```

display the images with subplot

```python
import matplotlib.pyplot as plt
```

```python
for i in range(8):
    plt.subplot(3, 3, i + 1)
    io.imshow(bit_img[i])
```
