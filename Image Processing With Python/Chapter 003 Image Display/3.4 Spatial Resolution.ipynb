## Reducing the resolution of an image

```python
import matplotlib.pyplot as plt
import skimage.io as io
import skimage.transform as tr
```

```python
img = io.imread('_images/apple.jpeg')
io.imshow(img)
```

```python
for i in range(8):
    n = 1 << i
    print(n)
    img1 = tr.rescale(img, 1/n)
    plt.subplot(3, 3, i + 1)
    io.imshow(img1)
```

## In a line

```python
imgs = [tr.rescale(img, 1/(1 << i)) for i in range(8)]
```

```python
io.imshow(imgs[4]) # resulation is 1/16
```

```python
io.imshow(imgs[6]) # resulation is 1/64
```
