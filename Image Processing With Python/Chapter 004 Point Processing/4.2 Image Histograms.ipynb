Given a grayscale image, its histogram consists of the histogram of its gray levels; that is, a graph
indicating the number of times each gray level occurs in the image. We can infer a great deal about the
appearance of an image from its histogram, as the following examples indicate:
<ul>
    <li>In a dark image, the gray levels (and hence the histogram) would be clustered at the lower end.</li>
    <li>In a uniformly bright image, the gray levels would be clustered at the upper end.</li>
    <li>In a well-contrasted image, the gray levels would be well spread out over much of the range.</li>
</ul>

```python
import skimage.io as io
```

```python
im = io.imread('_images/caribou.jpg')
io.imshow(im)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.show(plt.hist(im.flatten(), bins=256))
```

### Make the image darker

```python
im1 = im // 5
io.imshow(im1)
```

```python
plt.show(plt.hist(im1.flatten(), bins=256))
```

## Histogram Stretching, Adjustment

```python
import skimage.exposure as ex
```

```python
im2 = ex.adjust_gamma(im1, 0.4)

plt.subplot(2, 1, 1)
io.imshow(im2)
plt.subplot(2, 1, 2)
plt.show(plt.hist(im2.flatten(), bins=256))
```

```python
im2 = ex.adjust_gamma(im1, 1.2)

plt.subplot(2, 1, 1)
io.imshow(im2)
plt.subplot(2, 1, 2)
plt.show(plt.hist(im2.flatten(), bins=256))
```

## Histogram Equalization

```python
import skimage.exposure as ex
```

```python
im_heq = ex.equalize_hist(im1)
io.imshow(im_heq)
```

```python
plt.show(plt.hist(im_heq.flatten(), bins=256))
```
